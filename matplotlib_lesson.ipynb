## Example 1

```python
from random import randint

x1 = list(range(10))
y1 = [randint(0,10) + j for j in x1]

x2 = list(range(10))
y2 = [randint(0,10) + j for j in x2]

x3 = list(range(10))
y3 = [randint(0,10) + k for k in x3]
```

### Basic plot

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.show()
```

### Custom ticks & rotate

- .xticks()
- .xticks(rotation=)

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)

plt.xticks([0,2.5,3,6,10],
           ['zero','two point five', 'three', 'six', 'ten'],
          rotation=-60)

plt.show()
```

### Markers & line styles

- plot(marker=, ls=)

```python
x1, y1
```

```python
plt.plot(x1,y1, marker='o', ls='--')
plt.plot(x2,y2, ls=':', marker='^')
plt.plot(x3,y3)

plt.xticks([0,2.5,3,6,10],
           ['zero','two point five', 'three', 'six', 'ten'],
          rotation=-60)

plt.show()
```

### Figure & Subplots

- .figure(figsize=()), .subplot()

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.plot(x1,y1, marker='o', ls='--')

plt.subplot(132)
plt.plot(x2,y2, ls=':', marker='^')

plt.subplot(133)
plt.plot(x3,y3)

plt.show()
```

### Add limits

- .xlim(), .ylim()

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.plot(x1,y1, marker='o', ls='--')
plt.ylim(0,16)

plt.subplot(132)
plt.plot(x2,y2, ls=':', marker='^')
plt.ylim(0,16)

plt.subplot(133)
plt.plot(x3,y3)
plt.ylim(0,16)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(x1,y1, marker='o', ls='--')
plt.plot(x2,y2, ls=':', marker='^')
plt.plot(x3,y3)

plt.xlim(2,6)
plt.ylim(0,14)

plt.show()
```

### Add text, lines, & arrows!!

> .text, .hlines, .vlines, .annotate

```python
plt.figure(figsize=(12,6))

plt.plot(x1,y1, marker='o')
plt.plot(x2,y2, ls=':', marker='^')
plt.plot(x3,y3)

plt.hlines(11,0,3, ls='--')
plt.vlines(5,0,15, ls='--')

plt.text(1,11.5,'(1,11)')

plt.annotate('this is importat', 
             xy=(5.1,5.9),
             xytext=(5.5,4.5),
             arrowprops={'facecolor':'grey'})

plt.show()
```

## Example 2

data obtained from: https://www.kaggle.com/datasets/teertha/ushealthinsurancedataset

```python
import pandas as pd
```

```python
df = pd.read_csv('insurance.csv')
df
```

### Basic plot

- bmi vs charges

```python
plt.scatter(df.bmi,df.charges)
plt.show()
```

### Basic plot - two sets of data

- bmi vs charges for smoker and non-smoker

```python
df_smoke_no = df [df.smoker == 'no']
df_smoke_yes = df [df.smoker == 'yes']
```

```python
plt.scatter(df_smoke_no.bmi, df_smoke_no.charges)
plt.scatter(df_smoke_yes.bmi, df_smoke_yes.charges)
plt.show()
```

### Color

- list of colors in matplotlib: https://matplotlib.org/stable/gallery/color/named_colors.html

```python
plt.scatter(df_smoke_no.bmi, df_smoke_no.charges, color='blue')
plt.scatter(df_smoke_yes.bmi, df_smoke_yes.charges, color='orchid')
plt.show()
```

### Transparency

```python
plt.scatter(df_smoke_no.bmi, df_smoke_no.charges, 
            color='blue', alpha=.4)
plt.scatter(df_smoke_yes.bmi, df_smoke_yes.charges, 
            color='orchid', alpha=.4)
plt.show()
```

### Size

```python
plt.scatter(df_smoke_no.bmi, df_smoke_no.charges, 
            color='blue', alpha=.4, s=10, marker='^')
plt.scatter(df_smoke_yes.bmi, df_smoke_yes.charges, 
            color='orchid', alpha=.4, s=100, marker='*')
plt.show()
```

### Finalize - labels

- .figure, .title, .xlabel, .ylabel, .legend

```python
plt.figure(figsize=(12,6))

plt.scatter(df_smoke_no.bmi, df_smoke_no.charges, 
            color='blue', alpha=.5, s=10, marker='^')
plt.scatter(df_smoke_yes.bmi, df_smoke_yes.charges, 
            color='orchid', alpha=.5, s=200, marker='*')

plt.title('How do insurance charges change by bmi for smokers')
plt.xlabel('bmi')
plt.ylabel('insurance charges')
plt.legend(['non-smokers','smokers'])

plt.show()
```
