# 1. Copy the code from the lesson to create a dataframe full of student grades.

```python
# importing pandas for tabular manipulation
import pandas as pd
# importing numpy for vectorized math operations on the content
import numpy as np

# set a random seed to replicate results
np.random.seed(1349)

# define a list of students
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']
```

```python
from pydataset import data
```

```python
# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))
```

```python
# each piece of grade information is ordered and as long as the students
math_grades
```

```python
len(math_grades) == len(english_grades) == len(reading_grades) == len(students)
```

```python
df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
# big three investigative calls for a new dataframe:
# df.info()
#  -- basic details of the dataframe: size of df, glaringly obvious nulls (usually*), columns, datatypes
# df.head()
#  -- get the first few rows of the dataframe
# df.describe()
#  -- describe gives us our basic descriptive statistics
```

```python
df.describe()
```

```python
df.info()
```

### Create a column named passing_english that indicates whether each student has a passing grade in english.

```python
# how do we establish the truth value of whether or not a student is passing english?
```

```python
# we have to create a boolean mask that will apply to the values inside of 
# the Series 'english'

# apply this boolean mask to the actual values to get an array of Truth values

# assign this column into our dataframe
```

```python
df['english']
```

```python
df['english'] >= 70
```

```python
# assign the new value 
df['passing_english'] = df['english'] >= 70
```

```python
df.head()
```

### Sort the english grades by the passing_english column. How are duplicates handled?

```python
pd.DataFrame.sort_values?
```

```python
df.sort_values(by='passing_english')
```

* **Duplicates are handeled in ascending index order**

### Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.

```python
df.sort_values(by=['passing_english', 'english'], ascending=[True, False])
```

### Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

```python
# presume each category of grade is weighted equally
df['overall'] = round((df.math + df.english + df.reading) / 3, 0).astype(int)
```

```python
df.head(2)
```

```python
# how many students passed in english?
len(df[df.passing_english])
```

```python
# keep this in our pocket for later when we think about the total number of values
# that may or may not be null
df.passing_english.sum()
```

```python
# What proportion of students passed in english?

round(df.passing_english.sum() / len(df) * 100, 2)
```
